````python
b = np.array([0,8,8,20])
t = np.array([0,1,3,4])
````

````python
# Solve At = b using Pseudoinverse

one = np.ones(4)
A = np.vstack((one,t)).T

U,Sigma,VT = npl.svd(A)

S_dag = np.zeros(A.T.shape)
np.fill_diagonal(S_dag,1/Sigma)

A_dag = VT.T @ S_dag @ U.T

ans = A_dag @ b

print(ans)
````

````python
# Solving using the normal equations

npl.solve(A.T@A,A.T@b)
````

````python
print('The heights are : ',A@ans)
````

````python
print('The errors are : ',np.abs(A@ans - b))
````

````python
# Now solve for a line that goes through the origin using the normal equations

A = t

slope = (A.T @ b) / (A.T@A)
print(slope)
````

##### Question 2

````python
# Use PCA to find the line of best fit

A = np.vstack((t,b))

A[0,:] = A[0,:] - A[0,:].mean()
A[1,:] = A[1,:] - A[1,:].mean()

U,Sigma,VT = npl.svd(A)
S = np.zeros(A.shape)
np.fill_diagonal(S,Sigma)

direction_1 = U[:,0]
direction_2 = U[:,1]

# Eigenvalues which directions are most important 

npl.eig(A@A.T)
````

````python
plt.scatter(t,b)

slope1 = direction_1[1]/direction_1[0]
xs = np.linspace(0,4,10)
ys = b.mean() + slope1 * (xs - t.mean())

plt.plot(xs,ys)
````

````python
errors = A.T@direction_2
errors
````

````python
reg_data = np.genfromtxt("/Users/jacksherry/Downloads/regression.csv", delimiter=",")
````

````python
# Obtain a least squares fit to the data using Normal Equations

xs = reg_data[:,0]
ys = reg_data[:,1]
ones = np.ones(reg_data.shape[0])
fs = reg_data[:,2]

A = np.vstack((ones,xs,ys)).T

a,b,c = npl.solve(A.T@A,A.T@fs)
print(a,b,c)
````

````python
x_plot = np.linspace(-5,5,100)
y_plot = np.linspace(-5,5,100)

X,Y = np.meshgrid(x_plot,y_plot)
Z = a + b*X + c*Y

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

ax.plot_surface(X,Y,Z,cmap = 'viridis')
ax.scatter(xs,ys,fs)
````

````python
errors = np.abs(fs - a - b*xs - c*ys)
print('The squared error is %.2f'%(np.sum(errors**2)))
````

````python
# Solving using CVXPY

beta = cp.Variable(3)

obj = cp.Minimize(cp.sum(cp.square(A@beta - fs)))

prob = cp.Problem(obj)
prob.solve()
````

````python
# Now solving using PCA

pca_xs = xs - np.mean(xs)
pca_ys = ys - np.mean(ys)
pca_fs = fs - np.mean(fs)

pca_data = np.vstack((pca_xs,pca_ys,pca_fs)).T

U,Sigma,VT = npl.svd(pca_data)

direction1 = VT.T[:,0]
direction2 = VT.T[:,1]
direction3 = VT.T[:,2]

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

ax.scatter(xs,ys,fs)
````

````python
normalv = np.cross(direction1,direction2)
point = np.array([np.mean(xs),np.mean(ys),np.mean(fs)])

d = point@normalv

plotxs = np.linspace(-5,5,100)
plotys = np.linspace(-5,5,100)

X,Y = np.meshgrid(plotxs,plotys)
Z = (d - normalv[0]*X - normalv[1]*Y)/normalv[2]

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot_surface(X,Y,Z,cmap = 'viridis')
ax.scatter(xs,ys,fs)
````

````python
errors = np.abs(pca_data@direction3)
print('The squared error is %.2f'%np.sum(errors**2))
````

# PCA 2D Example

---

## 1. Data Generation

We generate a synthetic dataset with a linear relationship between $x$ and $y$, plus some added noise:
\[
y = 3 + 2x + \epsilon, \quad \epsilon \sim \mathcal{N}(0, 0.6)
\]
where $x$ is evenly spaced between 0 and 5. The dataset contains 1001 points.

---

### What Happens Here:
- A synthetic dataset is generated where:
  - $x$ values are evenly spaced between 0 and 5 (1001 points).
  - $y = 3 + 2x + \text{noise}$, where the noise is Gaussian ($\mu = 0$, $\sigma = 0.6$).
- This creates a dataset with a clear linear relationship between $x$ and $y$, with some random deviations.

### Purpose:
To provide a toy dataset to visualize how PCA works on data with a linear trend.

---

## 2. Data Centering (Standardization)

We center the data by subtracting the mean of $x$ and $y$:
\[
\tilde{x} = x - \bar{x}, \quad \tilde{y} = y - \bar{y}
\]

### What Happens Here:
- The mean is subtracted from $x$ and $y$ to center the data around the origin.
- This is essential for PCA because PCA relies on variance, and centering ensures that the principal components capture variance relative to the mean.

### Result:
- A 2D array `standardizedData` with centered $x$ and $y$.

---

## 3. Singular Value Decomposition (SVD)

We perform SVD on the standardized data:
\[
\text{standardizedData} = U \cdot S \cdot V^T
\]
where:
- $U$: Projections of the data onto the principal components.
- $S$: Diagonal matrix of singular values, representing the variance explained by each component.
- $V^T$: Principal component directions (eigenvectors of the covariance matrix).

---

### What Happens Here:
- **SVD Decomposition**:
  \[
  \text{standardizedData} = U \cdot S \cdot V^T
  \]
  - $U$: Contains the projections of the data onto the principal components.
  - $S$: Singular values (variance explained by each principal component).
  - $V^T$: Rows contain the principal directions (eigenvectors of the covariance matrix).
- **Diagonal Matrix $S$**:
  - The singular values $S$ are placed on the diagonal to facilitate reconstruction or projections.

### Purpose:
To extract the principal components and their variance (singular values).

---

## 4. Plotting Principal Directions

We plot the principal directions scaled by the singular values.

---

### What Happens Here:
- **Red Line (1st Principal Direction)**:
  - Plots the first principal direction scaled by the singular value $S_1$.
  - This represents the direction along which the variance is maximized.
- **White Line (2nd Principal Direction)**:
  - Plots the second principal direction scaled by the singular value $S_2$.
  - This is orthogonal to the first principal direction and captures the remaining variance.

### Purpose:
To visualize how PCA finds the two orthogonal directions that capture the variance in the data.

---

## 5. Projecting Down to 1D

The data is projected onto the first principal component:
\[
\text{Projection 1D: } \text{Projection} = U[:, 0] \cdot S_1
\]

---

### What Happens Here:
- Projects the data onto the first principal component (1D space).

### Purpose:
To reduce the dimensionality of the dataset while preserving the maximum variance.

---

## 6. Projecting Down to 2D

The data is reconstructed in the 2D space using the first two principal components:
\[
\text{Projection 2D: } U[:, :2] \cdot S_{1:2}
\]

---

### What Happens Here:
- Projects the data back into a 2D space.

### Purpose:
To reconstruct the dataset using only the first two principal components (useful for high-dimensional datasets).

---

## Key Outputs

### Singular Values:
The singular values represent the variance along the principal components:
\[
S = \begin{bmatrix}
103.71 & 0 \\
0 & 8.41
\end{bmatrix}
\]

### Visualizations:
1. **Original Data with Principal Directions**:
   - Shows the original dataset with the two orthogonal principal directions.
2. **1D Projection**:
   - Displays the dataset reduced to 1D using the first principal component.
3. **2D Projection**:
   - Reconstructed 2D data from the principal components.

---

## Summary

This example demonstrates PCA in 2D with the following steps:
1. **Generate Data**: Create a synthetic linear dataset with noise.
2. **Center Data**: Subtract the mean to center the data.
3. **Perform PCA**: Decompose the data into principal components using SVD.
4. **Visualize Principal Directions**: Plot the directions capturing variance.
5. **Projection**:
   - Reduce to 1D for maximum variance.
   - Reconstruct in 2D for visualization.


### PCA 3D Example

````python
[x,y,z] = np.array([[x,3 + 2*x + 0.6*np.random.randn(),7 + 4*x + 6*np.random.randn()]
                    for x in np.linspace(0,5,1001)]).T
````

````python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
````

````python
x = (x-np.mean(x))/np.std(x)
y = (y-np.mean(y))/np.std(y)
z = (z-np.mean(z))/np.std(z)
standardizedData = np.array([x,y,z]).T
````

````python
U,Svec,VT = npl.svd(standardizedData)
S = np.zeros(standardizedData.shape)
np.fill_diagonal(S,Svec)
Svec
````

````python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z, s=0.2)
# First principal component
ax.plot([0,Svec[0]*VT[0,0]/np.sqrt(x.size -1)],
         [0, Svec[0]*VT[0,1]/np.sqrt(x.size -1)],
         [0, Svec[0]*VT[0,2]/np.sqrt(x.size -1)],"r-")

# Second principal component
ax.plot([0,-Svec[1]*VT[1,0]/np.sqrt(x.size -1)],
         [0, -Svec[1]*VT[1,1]/np.sqrt(x.size -1)],
         [0, -Svec[1]*VT[1,2]/np.sqrt(x.size -1)],"b-")


# Third principal component
ax.plot([0,-Svec[2]*VT[2,0]/np.sqrt(x.size -1)],
         [0, -Svec[2]*VT[2,1]/np.sqrt(x.size -1)],
         [0, -Svec[2]*VT[2,2]/np.sqrt(x.size -1)],"k-")

plt.show()
````

````python
# Plot projected data
[x_proj, y_proj, z_proj] = (U[:,:3]@S[:3,:3]).T
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_proj, y_proj, z_proj)

plt.axis('equal')
plt.show()
````

### PCA Classification

````python
from sklearn.datasets import load_iris
iris = load_iris()
````

````python
# There are 4 features and 3 types of flowers (categories)

[iris.feature_names,iris.target_names]
````

````python
# Getting the data in the right format and then performing an SVD

data = iris.data
for i in range(0,4):
    data[:,i] = data[:,i] - np.mean(data[:,i])
    
U,Sigma,VT = npl.svd(data)
S = np.zeros(data.shape)
np.fill_diagonal(S,Sigma)
````

````python
# We will now project down a dimension and plot the data

[x_proj, y_proj, z_proj] = (U[:,:3]@S[:3,:3]).T
````

### Assignment 2 : 2D Case

````python
plus = np.array([[-1,2],[-3,3],[-4,3]])
minus = np.array([[3,-2],[3,-4],[4,-3]])
````

````python
# Solving the primal problem 

w = cp.Variable(2)
b = cp.Variable(1)

constraints = [plus@w + b >= 1 , minus@w + b <= -1]

obj = cp.Minimize(cp.sum(cp.square(w)))

prob = cp.Problem(obj,constraints)
prob.solve()

print(w.value,b.value)
````

````python
# Solving the dual problem

lam = cp.Variable(plus.shape[0]+minus.shape[0])

ys = np.ones(plus.shape[0]+minus.shape[0])
ys[plus.shape[0]:] = -1
X = np.vstack((plus,-minus))
e = np.ones(plus.shape[0]+minus.shape[0])

constraints = [ys@lam == 0 , lam >= 0]

obj = cp.Minimize(0.5 * cp.quad_form(lam,X@X.T,assume_PSD = True) - lam@e)

prob = cp.Problem(obj,constraints)
prob.solve()

w = X.T@lam.value
index = np.argmax(lam.value)
b = ys[index] - w@X[index]

print(w,b)
````

### Gradient Descent

````python
# Quadratic Optimisation Problem : f(x) = 0.5 xT A x - xT b

def Quad_GD(A,x,b,tol,maxit):
    steps = []
    funcvals = []
    steps.append(x)
    res = A@x - b 
    iter = 0
    converged = False
    while not converged and iter < maxit:
        gradx = A@x - b
        alpha = npl.norm(gradx)**2 / (gradx @ A @ gradx)
        x = x - alpha*gradx
        res = A@x - b
        steps.append(x)
        funcvals.append(0.5*x@A@x - x@b)
        
        iter += 1
        
    return steps,funcvals
````

````python
# We initialise b to be (3,3) which means that (1,1) is the minimum

A = np.array([[7,-4],[-4,7]])
b = np.array([3,3])
descent,fs = Quad_GD(A,[5,8],b,10e-4,10)
````

````python
xs = np.linspace(0,6,100)
ys = np.linspace(0,10,100)
X,Y = np.meshgrid(xs,ys)

Z = np.zeros(X.shape)

for i in range(X.shape[0]):
    for j in range (X.shape[1]):
        v = np.array([X[i,j],Y[i,j]])
        Z[i,j] = 0.5*v@A@v-v@b
        
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(2, 1, 1)
plt.contourf(X, Y, Z, levels=25)
plt.colorbar()

plotxs = np.zeros(10)
plotys = np.zeros(10)
plotfs = np.zeros(10)
for i in range(0,10):
    plotxs[i] = descent[i][0]
    plotys[i] = descent[i][1]
    plotfs[i] = fs[i]

ax.plot(plotxs,plotys,color = 'red')

ax = fig.add_subplot(2, 1, 2, projection='3d')
n = 150
ax.contour3D(X,Y,Z, n, alpha = .5)
ax.plot(plotxs,plotys,plotfs)
````

# Optimization: Gradient Descent on a Quadratic Function

This example demonstrates the implementation and visualization of gradient descent for optimizing a quadratic function.

---

## 1. Problem Description

The quadratic optimization problem is defined as:
\[
f(x) = \frac{1}{2} x^T A x - x^T b
\]
where:
- $A$ is a positive definite matrix, ensuring the problem has a unique minimum.
- $b$ is a vector that influences the position of the minimum.

The gradient of $f(x)$ is:
\[
\nabla f(x) = A x - b
\]

The optimal solution satisfies:
\[
A x = b \quad \Rightarrow \quad x = A^{-1} b
\]

---

## 2. Gradient Descent Algorithm

### Implementation:

The gradient descent algorithm iteratively updates $x$ to minimize $f(x)$ using the formula:
\[
x_{k+1} = x_k - \alpha_k \nabla f(x_k)
\]
where:
- $\alpha_k$ is the step size (adaptive in this case).
- $\nabla f(x_k) = A x_k - b$ is the gradient at iteration $k$.

The step size is computed as:
\[
\alpha_k = \frac{\|\nabla f(x_k)\|^2}{\nabla f(x_k)^T A \nabla f(x_k)}
\]

### Code:

```python
def Quad_GD(A, x, b, tol, maxit):
    steps = []
    funcvals = []
    steps.append(x)
    res = A @ x - b
    iter = 0
    converged = False
    while not converged and iter < maxit:
        gradx = A @ x - b
        alpha = npl.norm(gradx)**2 / (gradx @ A @ gradx)
        x = x - alpha * gradx
        res = A @ x - b
        steps.append(x)
        funcvals.append(0.5 * x @ A @ x - x @ b)
        iter += 1
    return steps, funcvals

### Gradient Descent with Momentum

````python
def Quad_GD_MMTM(A,x,b,alpha,beta,maxit,tol):
    steps = []
    funcvals = []
    steps.append(x)
    z = A@x - b
    iter = 0
    converged = False
    while not converged and iter < maxit:
        x = x - alpha*z
        res = A@x - b
        grad = A@x -b
        z = grad + beta*z
        steps.append(x)
        funcvals.append(0.5*x@A@x - x@b)
        iter += 1
            
    return steps,funcvals
````

The optimal choice for $\alpha$ and $\beta$ can be determined as
$$\alpha = \left(\frac{2}{\sqrt{\lambda_{max}(A)}+\sqrt{\lambda_{min}(A)}}\right)^2,
\qquad \beta = \left(\frac{\sqrt{\lambda_{max}(A)}-\sqrt{\lambda_{min}(A)}}{\sqrt{\lambda_{max}(A)}+\sqrt{\lambda_{min}(A)}}\right)^2$$

````python
evals = npl.eig(A)[0]
alpha = (2/(np.sqrt(evals[0]) + np.sqrt(evals[1])))**2
beta = ((np.sqrt(evals[0]) - np.sqrt(evals[1]))/(np.sqrt(evals[0]) - np.sqrt(evals[1])))
````

````python
A = np.array([[7,-4],[-4,7]])
b = np.array([3,3])
mmdescent,mmfs = Quad_GD_MMTM(A,[5,8],b,alpha,beta,10e-4,10)
````

# Gradient Descent with Momentum

Gradient Descent with Momentum improves the convergence of the standard gradient descent by adding a momentum term to the gradient update. This method helps accelerate convergence, particularly in scenarios where the cost function has a narrow valley structure.

---

## 1. Problem Setup

The optimization problem is:
\[
f(x) = \frac{1}{2} x^T A x - x^T b
\]
where:
- $A$ is a positive definite matrix.
- $b$ is a vector that defines the position of the minimum.

The minimum is given by:
\[
A x = b \quad \Rightarrow \quad x = A^{-1} b
\]

---

## 2. Gradient Descent with Momentum

### Algorithm:

The iterative update rule with momentum is:
\[
x_{k+1} = x_k - \alpha z_k
\]
\[
z_{k+1} = \nabla f(x_k) + \beta z_k
\]
where:
- $\alpha$ is the step size.
- $\beta$ is the momentum parameter.
- $z_k$ is a combination of the current gradient and the previous momentum term.

### Code:

```python
def Quad_GD_MMTM(A, x, b, alpha, beta, maxit, tol):
    steps = []
    funcvals = []
    steps.append(x)
    z = A @ x - b
    iter = 0
    converged = False
    while not converged and iter < maxit:
        x = x - alpha * z
        res = A @ x - b
        grad = A @ x - b
        z = grad + beta * z
        steps.append(x)
        funcvals.append(0.5 * x @ A @ x - x @ b)
        iter += 1
    return steps, funcvals
